```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import Input, Dense, Activation, Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.utils import timeseries_dataset_from_array
from tcn import TCN  # pip install keras-tcn
import matplotlib.pyplot as plt

# ==============================================================================
# 1. STANDARD TCN ARCHITECTURE
# ==============================================================================
def build_Standard_TCN(input_timepoints, input_chans=8, params=None):
    n_filters = params.get('NO_FILTERS', 32)
    n_kernels = params.get('NO_KERNELS', 4)
    n_dilations = params.get('NO_DILATIONS', 4)

    signal_input = Input(shape=(input_timepoints, input_chans))
    
    nets = TCN(nb_filters=n_filters, kernel_size=n_kernels, nb_stacks=1,
               dilations=[2 ** i for i in range(n_dilations)], padding='causal',
               use_skip_connections=True, return_sequences=True,
               activation='gelu', kernel_initializer='he_normal')(signal_input)
               
    # Extract the feature vector at the VERY LAST timepoint of the window
    feats = Lambda(lambda tt: tt[:, -1, :], name='extract_last_timepoint')(nets)

    cls_in = tf.stop_gradient(feats) if 'StopGrad' in params.get('TYPE_ARCH', '') else feats
    cls_logits = Dense(1, kernel_initializer='glorot_uniform', bias_initializer='zeros', name='cls_logits')(cls_in)
    cls_prob = Activation('sigmoid', name='cls_prob')(cls_logits)

    return Model(inputs=signal_input, outputs=cls_prob, name='tcn_standard')

# ==============================================================================
# 2. EXECUTION LOOP
# ==============================================================================
if __name__ == "__main__":
    TIMEPOINTS = 64
    CHANNELS = 8
    TOTAL_SAMPLES = 50000
    FS = 1250  
    
    print("1. Generating Continuous Dummy Data...")
    dummy_lfp = np.random.randn(TOTAL_SAMPLES, CHANNELS).astype(np.float32)
    dummy_labels = np.zeros(TOTAL_SAMPLES, dtype=np.float32)
    
    for _ in range(100):
        onset = np.random.randint(1000, TOTAL_SAMPLES - 1000)
        length = 20  
        dummy_labels[onset:onset+length] = 1.0
        
        t = np.arange(length)
        for ch in range(CHANNELS):
            dummy_lfp[onset:onset+length, ch] += 5.0 * np.hanning(length) * np.sin(2 * np.pi * 150 * t / FS)

    print("2. Creating Keras Timeseries Pipeline...")
    # Offset targets so window [0 to T] predicts label at T
    target_labels = dummy_labels[TIMEPOINTS - 1:]
    
    train_dataset = timeseries_dataset_from_array(
        data=dummy_lfp,
        targets=target_labels,
        sequence_length=TIMEPOINTS,
        sequence_stride=16,
        batch_size=128,
        shuffle=True
    )
    
    print("3. Compiling Model (Focal Loss)...")
    params = {'NO_FILTERS': 32, 'NO_KERNELS': 4, 'NO_DILATIONS': 4, 'TYPE_ARCH': ''}
    model = build_Standard_TCN(input_timepoints=TIMEPOINTS, input_chans=CHANNELS, params=params)
    
    loss_function = tf.keras.losses.BinaryFocalCrossentropy(apply_class_balancing=True, alpha=0.25, gamma=2.0)
    model.compile(optimizer=tf.keras.optimizers.AdamW(learning_rate=1e-3), loss=loss_function)

    print("4. Training...")
    model.fit(train_dataset, epochs=15)

    print("\n5. Inference & Plotting...")
    lfp_chunk = np.random.randn(5000, CHANNELS).astype(np.float32)
    label_chunk = np.zeros(5000, dtype=np.float32)
    
    for i in range(5):
        onset = 500 + i * 1000  
        length = 25  
        label_chunk[onset:onset+length] = 1.0
        t = np.arange(length)
        for ch in range(CHANNELS):
            lfp_chunk[onset:onset+length, ch] += 5.0 * np.hanning(length) * np.sin(2 * np.pi * 150 * t / FS)

    inference_dataset = timeseries_dataset_from_array(
        data=lfp_chunk, targets=None,
        sequence_length=TIMEPOINTS, sequence_stride=1,
        batch_size=256, shuffle=False
    )
    
    predicted_probs = model.predict(inference_dataset, verbose=0).flatten()
    
    # Pad at the beginning for causal alignment
    padded_probs = np.zeros(len(lfp_chunk))
    padded_probs[-len(predicted_probs):] = predicted_probs
    
    t_axis = np.arange(len(lfp_chunk)) / FS 
    fig, ax1 = plt.subplots(figsize=(15, 5))
    ax1.plot(t_axis, lfp_chunk[:, 0], color='black', alpha=0.5, label='LFP Signal')
    ax1.fill_between(t_axis, ax1.get_ylim()[0], ax1.get_ylim()[1], where=(label_chunk>0), color='green', alpha=0.3, label='Ripple GT')
    ax2 = ax1.twinx()
    ax2.plot(t_axis, padded_probs, color='red', linewidth=2.0, label='Confidence')
    ax2.set_ylim(-0.05, 1.05)
    fig.legend(loc='upper right', bbox_to_anchor=(0.9, 0.9))
    plt.title('Standard TCN (Focal Loss)')
    plt.show()
    
```
